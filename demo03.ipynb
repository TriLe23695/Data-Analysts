```python
'''
Xây dựng mô hình trí tuệ nhân tạo sử dụng Machine Learning
Khi xây dựng mô hình Machine Learning thì người ta thường chia tập dữ liệu thành 2 phần
1. Training data: tập huấn luyện cho mô hình
2. Testing data: Tập kiểm thử

Sinh viên cần hiểu được tại sao phải chia làm 2 tập (thường thì tập train sẽ nhiều hơn)
Từ đó hiểu được các khái niệm overfitting (học vẹt)
'''
```

```python
# Tập dữ liệu
X = df[['T5']].values
y = df[['T6']].values

# Chia thành 2 phần train và test, giả sử chọn tỉ lệ train là 0.8 và test là 0.2  (0.75 0.25)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
y_train
```

```python
from sklearn import linear_model

model=linear_model.LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
'''
T6 = 0.688*T5+2.416+epsilon
'''
```

```python
'''
Có rất nhiều phương pháp đánh giá
'''
print('Score: ',model.score(X_train,y_train))
print('Score: ',model.score(X_test,y_test))

'''
Trên 0.5 thì mô hình có khả năng sử dụng được
'''
```
